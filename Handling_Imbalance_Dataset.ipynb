## Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)  #1000*0.9=900
n_class_1 = n_samples - n_class_0           #1000 - 900 = 100
```

```python
n_class_0,n_class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0), #Most values will be between -1 and 1, but values beyond that are still possible.
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1), #Most values will be between 1 and 3, but values beyond that are still possible.
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
class_0
```

```python
class_1
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
print(df.to_string())
```

```python
df['target'].value_counts()
```

```python
## upsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_minority #there are 100 positive record
```

```python
df_majority   #there are 900 negative record
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, 
                                 replace=True,  # Sample with replacement means the data point can be selected multiple times
                                 n_samples=len(df_majority),  #The minority class is upsampled to have the same number of samples as the majority class.
                                 random_state=42  # Reproducibility for same op every time we run
                                )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

## Down Sampling

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_majority_downsampled=resample(df_majority,
                                 replace=True, 
                                 n_samples=len(df_minority),
                                 random_state=42
                                )
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_majority_downsampled,df_minority])
```

```python
df_downsampled['target'].value_counts()
```
